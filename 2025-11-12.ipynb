```python
def square_sum (n) :
    sum = 0
    for item in range(1,n + 1) :
       sum += item ** 2
    return sum
print(square_sum(3))
```

```python
list_1 = [1,2,3]
list_2 = [2,3,4]
def add_corresponding_numbers(list_1,list_2) :
    if len(list_1) != len(list_2) :
        return "Error: Lists should be in equal lenght"
    else:
        sum_list = []
        for item in range(len(list_1)) :
            sum_list.append(list_1[item] + list_2[item])
        return sum_list
print(add_corresponding_numbers(list_1,list_2))
```

```python
def calculate_total(price,quantity = 1,discount = 0) :
    total_cost = price*quantity
    discount_percentage = total_cost * discount/100
    total_cost -= total_cost * discount/100
    return "Total cost = ",total_cost,"Discount percentage =",discount_percentage


print(calculate_total(100,2,2))
print(calculate_total(1))
print(calculate_total(1,2))
```

```python
def sum_numbers(*args) :
    total = 0
    for i in args :
        total += i
    return total
print(sum_numbers(1,2,4))
```

```python
def average_positive(*args) :
    total = 0
    average = 0
    count = 0
    for i in args :
        if i > 0 :
            total += i
            count += 1
            average = total/count
    return average if count > 0  else "No positive numbers prived"
print(average_positive(-3,1,2))
```

```python
def average_score(**kwargs):
    total = 0
    count = 0
    for i in kwargs.values():
        total += i
        count += 1
    if count > 0:
        return total / count
    else:
        return "no subjects are provided"

print(average_score())                   
print(average_score(math=80, english=90))
```
