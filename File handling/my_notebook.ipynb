```python
file_1 = open("sample_1.txt","r")
#file path eka enter karanna ona
#methana ekama directory eke tiyena nisa name eka witarai danme
#r -> read mode
#default mode eka -> read 

#read karanawa
content = file_1.read()
print(content)
#object eka close karanawa
file_1.close()
```

```python
with open("sample_1.txt","r") as file:
    content = file.read()
    print(content)
#no need use close()
```

```python
with open("sample_1.txt","r") as file:
    #readline read one line
    line_1 = file.readline()
    print(line_1)
    line_2 = file.readline()
    print(line_2)
    line_3 = file.readline()
    print(line_3)
#no need use close()
#space enne \n charactor eka nisa
```

```python
with open("sample_1.txt","r") as file:
    #readline read one line
    line_1 = file.readline()
    print(line_1)
    line_2 = file.readline().strip()
    #strip() use karama \n eka ayin karanawa
    print(line_2)
    line_3 = file.readline()
    print(line_3)
#no need use close()
#space enne \n charactor eka nisa
```

```python
with open("sample_1.txt","r") as file:
    while True :
        content = file.readline().strip()
        if not content :
            break
        print(content)

#if not content  -> check  content eka twa tiyenawada
```

```python
with open("sample_1.txt","r") as file:
    line_1 = file.readlines()
    print(line_1)
#no need use close()
# readlines return list
```

```python
with open("sample_1.txt","r") as file:
    content = file.readlines()
    for item in content :
        print(item.strip())
#no need use close()
# readlines return list
```

```python
import csv
with open("person_data.csv","r",newline='') as file:
    content = csv.reader(file)
    for item in content:
        print(item)
#newline = '' scv waladi hariyta line ending handle karanna kyla csv libry ekata bara denawa
```

```python
import csv
with open("person_data.csv","r",newline='') as file:
    content = csv.DictReader(file)
    for item in content:
        print(item)
#newline = '' scv waladi hariyta line ending handle karanna kyla csv libry ekata bara denawa
```

```python
import csv 

csv_file_path = 'sample_data.csv'
data = [
    {"Name":"John Doe","Age":30,"City":"New York"},
    {"Name":"Jane smith","Age":30,"City":"Los Angeles"},
    {"Name":"Bob Johnson","Age":35,"City":"Chicago"}
]
fieldnames = ["Name","Age","City"]

with open(csv_file_path,'w',newline='') as csv_file:
    csv_writer = csv.DictWriter(csv_file,fieldnames=fieldnames)
    csv_writer.writeheader()
    csv_writer.writerows(data)
```

```python
#Q
#option 1
with open("input.txt", "r") as infile:
    content = infile.read().strip().split()

with open("even_lines.txt", "w") as outfile:
    for number in content:
        num = float(number)
        if num % 2 == 0:
            outfile.write(str(num) + "\n")
```

```python
#Q
#option 2
with open("input.txt", "r") as infile:
    content = infile.readlines()

with open("even_lines.txt", "w") as outfile:
    for i in range(1,len(content) +1,2):
        outfile.write(content[i])
```
