```python
class Player:
    score = 10

    #instance method(inside the class)
    def add_points(self,score):
       #methana self use karana nisa parameter eket danna ona self
       #self.score kiyanne me welawe hadan inna object eke score eka, score kiyala danne pitin ena value eka
       final_score = score + self.score
       return final_score
player_1 = Player()
print(player_1.score)
print(player_1.add_points(5))
```

```python
class Car:
    wheels = 4
    #hama object ekatama common attribute witarai udin danne
    
    def __init__(self,brand,color): #constructor
        self.brand = brand
        self.color = color
    def show_wheel_count(self) :
        return self.wheels
    
car_2 = Car("BMW","red")
print(f"{car_2.color} colour {car_2.brand} has {car_2.wheels} wheels")
```

```python
class Circle:
    def __init__(self,radious):
        self.radious = radious
    def area(self):
        return 3.14 * self.radious ** 2
    def circomference(self):
        return 2 * 3.14 * self.radious
    def display_info(self):
        #can call instance method inside instance method self.area()
        print(f"Circle with radius {self.radious} has area {self.area():.2f} and circomference {self.circomference():.2f}")

circle_1 = Circle(10)
circle_1.display_info()
```

```python
class Animal:
    def __init__(self,name):
        self.name = name
        
    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self,name,color):
        self.name = name
        self.color = color

    def speak(self):
        print(f"{self.name} says woof!.It is {self.color} in color")
dog_1 = Dog("buddy","white")
dog_1.speak()
```

```python
class Animal:
    def __init__(self,name):
        self.name = name
        
    def speak(self):
        print(f"{self.name} makes a sound")
class Dog(Animal):
    def __init__(self,name,color):
        super().__init__(name) #repeative widiyata code liyn netuwa kelinma super eke constructor eka use kara
        self.color = color
    def speak(self):
        print(f"{self.name} says woof!.It is {self.color} in color")

dog_1 = Dog("buddy","white")
dog_1.speak()
```

```python
from math import pi

class Shape:
    name = "shape"
    def __init__(self,name):
        self.name = name

    def area(self):
        print(F"Area calculation not defined for {self.name}")

class Square(Shape):
    def __init__(self,name,height,width):
        super().__init__(name)
        self.height = height
        self.width = width

    def area(self):
        area = self.width * self.height
        print(f"{self.name} area is {area}")
    
class Circle(Shape):
    def __init__(self,name,radious):
        super().__init__(name)
        self.radious = radious

    def area(self):
        area = pi * self.radious ** 2
        print(f"{self.name} area is {area}")

    
squre_1 = Square("square",23,34)
squre_1.area()
circle_1 = Circle("Circle",23)
circle_1.area()
```
