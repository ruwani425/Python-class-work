```python
#protected
class Square :
    def __init__(self,name):
        self._name = name
    def _area(self,side_length):
        return side_length * side_length

square =  Square("square")
print(square._name)
print(square._area(23))
```

```python

#private
class Square :
    def __init__(self,name):
        #private attribute
        self.__name = name

    #_ _ private
    def __area(self,side_length):
        return side_length * side_length

    #public
    def get_name(self) :
        return self.__name
    
    #public
    def set_name(self,name) :
        self.__name = name

    #public
    def get_area(self,side) :
        return self.__area(side)
    
square = Square("Square")
print(square.get_name())
square.set_name("sq")
print(square.get_name())
print(square.get_area(23))
```
