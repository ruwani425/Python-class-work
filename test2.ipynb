```python
numbers = [1,2,3,4]
match numbers :
    case [x] :
        print(x)
    case [x,y] :
        print(x*y)
    case [x,y,z] :
        print(x*y*z)
    case [x,y,z,a] :
        print(x*y*z*a)
    case _ :
        print("no answer")
```

```python
def get_stats(numbers):
    return min(numbers), max(numbers), sum(numbers) / len(numbers)
stats = get_stats([1, 9, 10])
print(type(stats))
```

```python
def summarize(*args):
    if len(args) == 0:
        return 0
    return sum(args) / len(args)

print(summarize(10, 20, 30))
```
