```python
a = 1
match a :
    case 1 :
        print("1")
    case 3 :
        print("3")
    case _:
        print("no matches value")
```

```python
numbers = [1,2,3,4]
match numbers :
    case [x] :
        print(x)
    case [x,y] :
        print(x*y)
    case [x,y,z] :
        print(x*y*z)
    case [x,y,z,a] :
        print(x*y*z*a)
    case _ :
        print("no answer")
```
